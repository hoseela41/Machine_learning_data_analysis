## Dimension reduction using PCA on datasets

## diabetes

```python
dims_diabetes = list(range(1,diabetesX_df.shape[1] + 1,1))
mse_list = []
final_ratio = None
for dim in dims_diabetes:
    pca = PCA(n_components = dim, random_state=random_seed)
    pca.fit_transform(diabetesX_df.values)
    rebuild = np.dot(pca.fit_transform(diabetesX_df.values), np.linalg.pinv(pca.components_.T))
    mse = mean_squared_error(diabetesX_df.values, rebuild)    
    mse_list.append(mse)
    final_ratio = pca.explained_variance_ratio_
#     print(dim, ": ", sum(pca.explained_variance_ratio_), ", ", mse, pca.explained_variance_ratio_)

algo = np.arange(1, diabetesX_df.shape[1] + 1, 1) 
width = 0.35      
labels = algo
x = np.arange(len(labels))  # the label locations

fig = plt.figure()
ax = fig.add_subplot(111)
rects2 = ax.bar(x, final_ratio, width, label='EigenValue')
ax.set_ylim(bottom=0.0, top=0.5)
ax2 = ax.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:red'
ax2.set_ylabel('Mean square errors', color=color)  # we already handled the x-label with ax1
ax2.plot(x, mse_list, color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Eigen Value')
ax.set_title('PCA on Diabetes')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
ax.grid(color='gray', linestyle='-', linewidth=0.1)

fig.tight_layout()
```

```python
comp = 7
all_pca = PCA(n_components = comp, random_state = random_seed)
pca_results = all_pca.fit_transform(diabetesX_df.values)
for i in range(comp):
    key_name = "PC"+str(i + 1)
    diabetesX_df[key_name] = pca_results[:,i]
    diabetes_dataset[key_name] = pca_results[:,i]
print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))
diabetes_pca_7 = pca_results
```

## Dimension reduction using PCA, clustering using K-means

## diabetes

```python
from sklearn.metrics import accuracy_score, homogeneity_score, completeness_score
clusters_range = list(range(2,9,1))
SE = []
silh = []
homog = []
CS = []
for cluster in clusters_range:
    kmeans = KMeans(n_clusters=cluster, max_iter=1000, random_state=random_seed, 
                    n_jobs=-2, algorithm = "auto").fit(diabetes_pca_7)
    k_label = kmeans.labels_
    sil_coeff = silhouette_score(diabetes_pca_7, k_label, metric='euclidean')
    homog_score = homogeneity_score(diabetesY, k_label)
    com_score = completeness_score(diabetesY, k_label)
    silh.append(sil_coeff)
    homog.append(homog_score)
    SE.append(kmeans.inertia_)
    CS.append(com_score)
    
# ========== plotting
# Create some mock data
fig = plt.figure()
ax1 = fig.add_subplot(111)
labels_1 = ['Squared Errors']
labels_2 = ["Silhouette Coefficient", "Homogeneity Score", "Completeness Score"]

color = 'black'
ax1.set_xlabel('Number of cluster')
ax1.set_ylabel('Squared Errors', color="blue")
ln1 = ax1.plot(clusters_range, SE, color="blue")
ax1.tick_params(axis='y', labelcolor="blue")

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:red'
ax2.set_ylabel('Score', color=color)  # we already handled the x-label with ax1
ln2 = ax2.plot(clusters_range, silh, color="orange")
ln3 = ax2.plot(clusters_range, homog, color=color)
ln4 = ax2.plot(clusters_range, CS, color="pink")
ax2.tick_params(axis='y', labelcolor=color)

ax1.set_title("Errors & score vs number of clusters for Diabetes (PCA + K-means)")
ax1.legend(labels_1, loc = 2)
ax2.legend(labels_2, loc = 7)
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

## Dimension reduction using PCA, clustering using EM

## diabetes

```python
clusters_range = list(range(2,9,1))
log_likelihood_EM = []
silh_EM = []
homog_EM = []
CS = []
for cluster in clusters_range:
    gmm = GaussianMixture(n_components=cluster, max_iter=100, random_state = random_seed, 
                          n_init=5).fit(diabetes_pca_7)
    gmm_label = gmm.predict(diabetes_pca_7)
    sil_coeff = silhouette_score(diabetes_pca_7, gmm_label, metric='euclidean')
    homog_score = homogeneity_score(diabetesY, gmm_label)
    comp_score = completeness_score(diabetesY, gmm_label)
    silh_EM.append(sil_coeff)
    homog_EM.append(homog_score)
    CS.append(comp_score)
    log_likelihood_EM.append(gmm.score(diabetes_pca_7))
    print("the calculation progress is: " + str(int(cluster / clusters_range[-1] * 100)) + "%", end = "\r")
    
    
# ========== plotting
# Create some mock data
fig = plt.figure()
ax1 = fig.add_subplot(111)
labels_1 = ['Squared Errors']
labels_2 = ["Silhouette Coefficient", "Homogeneity Score", "Completeness Score"]

color = 'black'
ax1.set_xlabel('Number of cluster')
ax1.set_ylabel('Log_likelihood', color="blue")
ln1 = ax1.plot(clusters_range, log_likelihood_EM, color="blue")
ax1.tick_params(axis='y', labelcolor="blue")

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:red'
ax2.set_ylabel('Score', color=color)  # we already handled the x-label with ax1
ln2 = ax2.plot(clusters_range, silh_EM, color="orange")
ln3 = ax2.plot(clusters_range, homog_EM, color=color)
ln4 = ax2.plot(clusters_range, CS, color="pink")
ax2.tick_params(axis='y', labelcolor=color)

ax1.set_title("Errors & score vs number of clusters for Diabetes (PCA + EM)")
ax1.legend(labels_1, loc = 2)
ax2.legend(labels_2, loc = 7)
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
